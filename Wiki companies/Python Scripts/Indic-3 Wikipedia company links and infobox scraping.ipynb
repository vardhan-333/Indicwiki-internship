```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://en.wikipedia.org/wiki/List_of_companies_of_India")
table= driver.find_elements(by = By.XPATH,value="//table[@class='wikitable sortable jquery-tablesorter'][2]/tbody/tr/td")
table
```

```python
cols=["Name","Industry","Sector","Headquarters","Founded","Notes","link"]
data=pd.DataFrame(columns=cols)
d={}
ctr=0
for val in table:
    d[cols[ctr]]=val.text
    if ctr==0:
        link=val.find_elements_by_tag_name("a")
        for x in link:
            d["link"]=x.get_attribute("href")
    ctr+=1
    if ctr==6:
        data=data.append(d,ignore_index=True)
        d={}
        ctr=0
data
```

```python
data.to_csv("D:/wiki.csv")
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
df=pd.read_csv("D:/wiki.csv")
ctr=0
d={}
data=pd.DataFrame()
for link in df["link"]:
    driver.get(link)
    table= driver.find_elements(by = By.XPATH,value="//table[@class='infobox vcard']/tbody/tr")
    for val in table:
        header=val.find_elements_by_tag_name("th")
        td=val.find_elements_by_tag_name("td")
        if header!=[] and td!=[]:
            d[header[0].text]=td[0].text
    data=data.append(d,ignore_index=True)
    d={}
    ctr+=1
data
```

```python
data.to_csv("D:/wiki_data.csv")
```

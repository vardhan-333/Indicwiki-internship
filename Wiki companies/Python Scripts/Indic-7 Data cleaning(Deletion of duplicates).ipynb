```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
ctr=0
for link in df["link"]:
    driver.get(link)
    table= driver.find_elements(by = By.XPATH,value="//table[@class='infobox vcard']/tbody/tr")
    for val in table:
        header=val.find_elements_by_tag_name("th")
        td=val.find_elements_by_tag_name("td")
        for h in header:
            if str(h.text).strip()=="Traded as":
                df.drop(df[df["link"]==link].index,inplace=True)
                ctr+=1
print(ctr)
df
```

```python
df.to_excel("C:/Users/bharg/Downloads/fil_wiki.xlsx")
```

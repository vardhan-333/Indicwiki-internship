```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
ctr=0
cols = ['CIN','Company Name','Link','RoC','Company Status']
data_frame = pd.DataFrame(columns=cols)
n=0
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
for page in range(8341,8500):
    driver.get('https://www.zaubacorp.com/company-list/p-'+str(page)+'-company.html')
    tab= driver.find_elements(by = By.XPATH,value="//table[@class='table table-striped col-md-12 col-sm-12 col-xs-12']/tbody/tr/td")
    l=[]
    for val in tab:
        if len(l)==4:
            l.append(val.text) 
            data_frame.loc[ctr] = l
            ctr+=1
            l=[]
        else:
            l.append(val.text)
            if len(l)==2:
                link = driver.find_element_by_link_text(val.text)
                l.append(link.get_attribute('href'))
print(ctr)
data_frame
```
